```python
import pandas as pd 
import numpy as np 
import ast
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies= movies.merge(credits,on='title')
```

```python
# genres , id, keyword , title
#overview, cast , crew 
# shief ye column bachenge data m 
movie=movies[['movie_id','genres','title','overview','cast','crew','keywords']]
```

```python
#missing data  check krna 
movie.isnull().sum()
```

```python
movie = movie.copy()
movie.dropna(subset=['overview'], inplace=True)
#overview column ki missing value ko drop kiya 
```

```python
movie.isnull().sum()
```

```python
movie.iloc[0].genres
```

```python
def covert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
 
```

```python
movie['genres']=movie['genres'].apply(covert)
```

```python
movie['keywords']=movie['keywords'].apply(covert)
```

```python
movie.head(1)
```

```python
def covert2(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
# fixing crew column 
```

```python
movie['cast']=movies['cast'].apply(covert2)
```

```python
movie['crew']
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
        
    return L
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie.head(1)
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
#sam ki problem htane ko transformation 
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tag']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
movie.head()
```

```python
new_df=movie[['movie_id','title','tag']]
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag']= new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))  # stem karke list me add karo
    return " ".join(y)        # list ko string me convert karke return
```

```python
ps.stem('loved')
```

```python
new_df['tag'].apply(stem)
```

```python
#calulating the distance between the movies
#csione distance measure krenge
# tetha between two vectors 
from sklearn.metrics.pairwise import cosine_similarity 
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list =sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
   
```

```python
recommend('Batman Begins')
```
